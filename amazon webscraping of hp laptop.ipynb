```python
cookie={}
```

```python
def getAmazonSearch(search_query):
    url="https://www.amazon.in/s?k="+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def urlSearch(search_query):
    url="https://www.amazon.in/dp/"+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
search_response=getAmazonSearch("hp+laptops")
```

```python
def r(search_query):
    url="https://www.amazon.in"+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
product=[]
for i in range(50):
    search_response=getAmazonSearch("hp+laptops&page="+str(i))
    soup=BeautifulSoup(search_response.content)
    for j in soup.findAll("span",attrs={'class':'a-size-medium a-color-base a-text-normal'}):
        product.append(j.text)
```

```python
tag=[]
for i in range(50):
    search_response=getAmazonSearch("hp+laptops&page="+str(i))
    soup=BeautifulSoup(search_response.content)
    for j in soup.findAll("div",attrs={'class':'sg-col-20-of-24 s-result-item sg-col-0-of-12 sg-col-28-of-32 sg-col-16-of-20 sg-col sg-col-32-of-36 sg-col-12-of-16 sg-col-24-of-28'}):
        tag.append(j['data-asin'])
        
```

```python
link=[]
for i in range(len(tag[i])):
    ##if tag[i]!="":
    response=urlSearch(tag[i])
    soup=BeautifulSoup(response.content)
    for j in soup.findAll('a',attrs={"data-hook":"see-all-reviews-link-foot"}):
        link.append(j['href'])
        
```

```python
link
```

```python
reviews=[]
for i in range(len(link)):
    response=r(link[i])
    soup=BeautifulSoup(response.content)
    for j in soup.findAll('span',attrs={"data-hook":"review-body"}):
        reviews.append(j.text)
```

```python
reviews
```

```python
len(reviews)
```

```python
reviews=[]
for i in range(len(link)):
    for k in range(1,50):
        response=r(link[i]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for j in soup.findAll('span',attrs={"data-hook":"review-body"}):
            reviews.append(j.text)
```
